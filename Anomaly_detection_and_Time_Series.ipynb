# Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

# Answer:

```
Anomaly Detection is the process of identifying rare or unusual data points that differ significantly from the majority of the dataset.

Types of Anomalies:

1. Point Anomaly
A single data point that is significantly different from others.
Example: A temperature sensor showing 120°C when normal range is 25°C.

2. Contextual Anomaly
An anomaly depending on context such as time or location.
Example: 30°C temperature is normal in summer but abnormal in winter.

3. Collective Anomaly
A group of data points that collectively behave abnormally.
Example: Sudden spike in network traffic for 10 minutes indicating cyber attack.
```

# Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

# Answer:

```
Isolation Forest
Approach: Randomly isolates data points using trees
Use Case: Large datasets and high dimensional data
Advantages: Fast and scalable

DBSCAN
Approach: Density based clustering
Use Case: Spatial data and irregular clusters
Advantages: Detects arbitrary shape clusters

Local Outlier Factor (LOF)
Approach: Density comparison with neighbors
Use Case: Local anomaly detection
Advantages: Good for detecting local anomalies
```

# Question 3: What are the key components of a Time Series? Explain each with one example.

# Answer:

```
Trend
Long-term upward or downward movement
Example: Increasing sales over years

Seasonality
Repeated pattern over fixed period

Cyclic

````
# Question 6

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Plot original
plt.figure(figsize=(10,5))
plt.plot(df)
plt.title("AirPassengers Data")
plt.show()

# Decompose
result = seasonal_decompose(df, model='multiplicative')
result.plot()
plt.show()

````

````
# Question 7: Apply Isolation Forest

# Answer:


# Question 7

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Sample dataset
np.random.seed(42)
data = np.random.randn(200,2)

df = pd.DataFrame(data, columns=['x','y'])

model = IsolationForest(contamination=0.05)
df['anomaly'] = model.fit_predict(df)

plt.scatter(df['x'], df['y'], c=df['anomaly'])
plt.title("Isolation Forest Anomalies")
plt.show()
````

````
# Question 8: Train SARIMA model

# Answer:

# Question 8

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

model = SARIMAX(df, order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit()

forecast = results.forecast(12)

plt.plot(df)
plt.plot(forecast)
plt.show()
````

````

# Question 9: Local Outlier Factor

# Answer:

# Question 9

from sklearn.neighbors import LocalOutlierFactor
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.random.randn(200,2)

lof = LocalOutlierFactor(n_neighbors=20)
y_pred = lof.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.title("LOF Anomaly Detection")
plt.show()
````
